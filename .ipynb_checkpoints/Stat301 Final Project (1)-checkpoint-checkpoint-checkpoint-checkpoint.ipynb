# (2) Introduction

The restaurant industry is highly competitive, with revenue influenced by various factors such as customer volume, pricing strategies, and marketing efforts. Understanding these relationships is crucial for business owners to optimize operations and maximize profits. 

Existing studies have explored how individual predictors are related to restaurant revenue. For example, Chen & Kim (2020) found that promotions could increase foot traffic (+18%) but would often erode long-term profitability (−7%). In a 2022 paper, Lee & Patel noted that dynamic menu pricing (e.g., surge pricing during peak hours) boosted revenue by 12% but risked customer dissatisfaction.

While these studies have examined individual factors affecting restaurant performance (Smith et al., 2023), few have explored their combined predictive power using simulated data to avoid real-world noise. Therefore, our research question is: Can we use predictors given in the simulated dataset (i.e., Number_of_Customers, Menu_Price, Marketing_Spend, Cuisine_Type, Average_Customer_Spending, and Promotions) to predict monthly restaurant revenue?

## c) Methods: Plan

We used additive multiple linear regression to predict because: 
1) The response (Monthly_Revenue) is continuous.

2) Our goal is to predict revenue based on the predictors given in the dataset.


Here are some assumptions for using multiple linear regression.

Assumptions: 
1) Linearity: The relationship between the response variable and the input variables should be linear in the coefficients. This means the effect of each input variable on the outcome variable should be additive and proportional.
2) Independence: Observations must be independent of one another. For example, there should be no correlation between residuals of different observations.
3) Homoscedasticity: The variance of residuals (errors) should remain constant across all levels of the independent variables.
4) Normality of Residuals: Residuals should be approximately normally distributed (crucial when performing hypothesis tests).
5) Multicollinearity: Input variables should not be highly correlated with each other. High multicollinearity can make it difficult to estimate the effect of each input variable accurately.
6) Model Specification: The model should include all relevant input variables and exclude irrelevant ones. Omitted variables or inclusion of unnecessary variables can bias results or increase variance.

Potential limitations:
1) Some of the assumptions above could be violated. For example, could the relationship be better accounted for by a nonlinear model?
2) Multiple linear regression either includes or excludes an input variable, but what if shrinking an input variable could produce better results?
3) MLR models are built on the data provided during training and assume that the relationships hold within that range. When one tries to extrapolate (make predictions outside the observed range of predictors), the model's predictions can become highly unreliable because it assumes the same linear relationship applies beyond the range of the data.

Meanwhile, we used **Backward Stepwise Selection** (via the `stepAIC` function from the MASS package) to help filter predictors. We chose `stepAIC` for the following reasons:
1. It ensures that no levels of dummy variables (such as Cuisine Type and Promotions) are discarded if those variables are selected.
2. It balances **model simplicity** (fewer variables) and **predictive power** (higher accuracy).
3. AIC is more interpretable than Adjusted R² for comparing models.

We opted for **backward stepwise selection** because:
1. The starting model was already simple, containing only six predictors: Number_of_Customers, Menu_Price, Marketing_Spend, Cuisine_Type, Average_Customer_Spending, and Promotions.
2. Forward selection risks excluding potentially useful variables, whereas backward selection preserves variables that may become significant after removing others.
3. Multicollinearity was low (as indicated by the previous heatmap), so removing variables wasn’t an immediate concern.

Steps Taken:
1. We started with a full model that included all predictors except Reviews.
2. Non-significant variables were removed using AIC to refine the model.

Alternative Methods Considered:
1. **LASSO Regression**: While LASSO is an effective method, it was not suitable for our small dataset with low multicollinearity.

To train our model, we first split the data into training and testing sets. Since we wanted more data for training, we used a **60-40 split** (rather than the more common 50-50 split). We performed the split using `tidymodels::initial_split()` and stratified the data by **Monthly_Revenue**.

```r
# separate the training data from the testing data
revenue_split <- 
    revenue_new %>%
    initial_split(prop = 0.6, strata = Monthly_Revenue) 

training_revenue <- training(revenue_split)
testing_revenue <- testing(revenue_split)
```

```r
# this is an additive model trained with training data
mlr_full <- lm(Monthly_Revenue~., data = training_revenue)
mlr_full
```

```r
# backward model selection using stepwise algorithm because:
# 1) we have two categorical variables
# 2) we don't have too many predictors, so it might be better to eliminate any unnecessary ones rather than adding
n <- nrow(training_revenue) # used to generate the penalty term, note that n is the number of observations
AIC_back <- stepAIC(mlr_full, direction = "backward", k = log(n))
summary(AIC_back)
```

```r
tidy(AIC_back)
```

It seems that only Number_of_Customers, Menu_Price, and Marketing_Spend are selected as predictors by the backward stepwise algorithm. We can also see that the coefficients on these predictors are all significant.

```r
# predicitons not made yet, use the test set. Haven't explained how to create the test set!!!

test_predictions <- predict(AIC_back, newdata = testing_revenue)

prediction_df <- data.frame(
  Actual = testing_revenue$Monthly_Revenue,
  Predicted = test_predictions)

metrics(data = prediction_df, truth = Actual, estimate = Predicted)
```

```r
ggplot(prediction_df, aes(x = Actual, y = Predicted)) +
  geom_point(alpha = 0.7, color = "steelblue") +
  geom_abline(intercept = 0, slope = 1, linetype = "dashed", color = "red") +
  labs(
    title = "Graph 3: Actual vs. Predicted Monthly Revenue (Test Set)",
    x = "Actual Monthly Revenue",
    y = "Predicted Monthly Revenue"
  ) +
  theme_minimal()
```

The final model demonstrates strong predictive performance on the test set, as shown by the close clustering of predicted revenue values around the actual values in the plot above. The R-squared, RMSE, and MAE metrics indicate that the model explains a substantial portion of revenue variation (As you can also see from the graph above, the points follow the red line clearly.). Most importantly, the model generalizes well to unseen data, supporting its validity for predicting restaurant revenue.

Additional notes:

# (4) Discussion

Our final multiple linear regression model identified **Number of Customers**, **Menu Price**, and **Marketing Spend** as the most important predictors of restaurant monthly revenue. These variables were statistically significant and explained a substantial portion of revenue variation, with an adjusted R² of approximately 0.68 on the training set. When applied to the test set, the model showed good predictive performance with a relatively small error and a strong alignment between actual and predicted values.

These results highlight the key drivers of revenue in a restaurant setting. As expected, attracting more customers and increasing average pricing are associated with higher revenue. Additionally, marketing efforts appear to play a strong role in boosting revenue, validating the business case for sustained promotional spending. Interestingly, **Promotions**, **Cuisine Type**, and **Average Customer Spending** were not retained in the final model, possibly due to redundancy or lack of independent predictive power.

Some limitations of this analysis include the use of **simulated data**, which may not fully capture the complexity of real-world restaurant operations. For instance, the presence of negative revenue values and the exclusion of operational costs or regional factors may oversimplify reality. Moreover, the model assumes linearity and additivity, which may not hold for all predictor relationships.

Future work could involve testing **non-linear models** (e.g., decision trees, random forests), adding **interaction terms** (e.g., Promotions × Marketing Spend), comparing our model trained with forward selection or LASSO, using cross-validation for training and testing, or analyzing **time-series data** to model trends and seasonality. Real-world applications could also benefit from a richer dataset that includes customer demographics, reviews sentiment, and competitive landscape.
